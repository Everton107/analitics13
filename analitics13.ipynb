```python
import pandas as pd

import datetime

from datetime import date, timedelta

import plotly.graph_objects as go

import plotly.express as px
```

```python
hoje = date.today()
d1 = hoje.strftime("%Y-%m-%d")

end_date = d1
d2 = date.today() - timedelta(days=365)

d2 = d2.strftime("%Y-%m-%d")

start_date = d2
```

```python
data = yf.download('AMZN',

                      start=start_date,

                      end=end_date,

                      progress=False)

data["Date"] = data.index
```

```python
#Escolhendo os tipos de dados para exibir

data = data[["Date", "Open", "High", "Low",

             "Close", "Adj Close", "Volume"]]

data.reset_index(drop=True, inplace=True)

#Imprimindo prévia

print(data.head())
```

```python
figure = go.Figure(data=[go.Candlestick(x=data["Date"],

                                        open=data["Open"], high=data["High"],

                                        low=data["Low"], close=data["Close"])])

figure.update_layout(title = "Análise de Ações da Amazon", xaxis_rangeslider_visible=False)

figure.show()
```

```python
figure = px.bar(data, x = "Date", y= "Close")

figure.show()
```

```python
figure = px.line(data, x='Date', y='Close',

                 title='Análise interativa do valor da ação')

figure.update_xaxes(rangeslider_visible=True)

figure.show()
```

```python
figure = px.line(data, x='Date', y='Close',

                 title='Análise de ações com períodos selecionados')



figure.update_xaxes(

    rangeselector=dict(

        buttons=list([

            dict(count=1, label="1m", step="month", stepmode="backward"),

            dict(count=6, label="6m", step="month", stepmode="backward"),

            dict(count=3, label="3m", step="month", stepmode="backward"),

            dict(count=1, label="1y", step="year", stepmode="backward"),

            dict(step="all")

        ])

    )

)

figure.show()
```

```python
figure = px.scatter(data, x='Date', y='Close', range_x=[d1, d2],

                 title="Análise do período definido sem fins de semanas")

figure.update_xaxes(

    rangebreaks=[

        dict(bounds=["sat", "sun"])

    ]

)

figure.show()
```
